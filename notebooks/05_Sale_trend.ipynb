```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("../data/clean/clean_data.csv")
```

```python
delivered_per_year = df.groupby("Year")["Delivered"].sum()
df[]

plt.figure(figsize=(8,5))
plt.bar(delivered_per_year.index, delivered_per_year.values)

plt.xlabel("Year")
plt.ylabel("Number of Delivered Orders")
plt.title("Delivered Orders per Year")

plt.xticks(delivered_per_year.index)
plt.show()
```
