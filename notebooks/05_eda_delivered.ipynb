```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("../data/clean/clean_data.csv")
```

```python
avg_revenue_per_day = (
    df[(df["Year"] == 2023) &
       (df["Shipping Status"] == "Delivered")]
    .groupby("Day")["Total Price"]
    .mean()
    .sort_values(ascending=False)
)

delivered_2023_per_month= (
    df[(df["Year"] == 2023) & (df["Delivered"] == True)]
    .groupby("Month_name")["Quantity"]
    .mean()
    .sort_values(ascending=False)
)
```

```python
# รวมเป็นระดับ Order ก่อน
order_level = (
    df[(df["Year"] == 2023) &
       (df["Shipping Status"] == "Delivered")]
    .groupby(["OrderID", "Month_name"])["Quantity"]
    .sum()
    .reset_index()
)

avg_qty_per_month = (
    order_level
    .groupby("Month_name")["Quantity"]
    .mean()
    .sort_values(ascending=False)
)
```

### 1) วันที่ มีค่าเฉลี่ยการ จัดส่ง(รับเงิน) มากที่สุด ของทั้งปี

```python
# Calculate average quantity of delivered orders per day
print("Average Quantity of Delivered Orders per Day:")
print(avg_qty_per_month)
```

##### ของทั้งปี 2023 พบว่า วันที่มีการสั่งซื้อ มากที่สุด ของทั้งปี 5อันดับ เรียงจากมากไปน้อย คือ (15, 24, 17, 3, 19)

### 1) เดือนที่ มีค่าเฉลี่ยการ จัดส่ง(รับเงิน) มากที่สุด ของทั้งปี

```python
# Calculate average quantity of delivered orders per month
print("Average Quantity of Delivered Orders per Month:")
print(delivered_2023_per_month)
```

##### ของทั้งปี 2023 พบว่า เดือน ที่มีการสั่งซื้อ มากที่สุด ของทั้งปี 5อันดับ เรียงจากมากไปน้อย คือ ( มีนาคม, มกราคม, กันยายยน, ตุลาคม, ธันวาคม)
