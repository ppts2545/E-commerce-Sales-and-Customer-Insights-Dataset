```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("../data/clean/clean_data.csv")
```

```python
# Convert "Date" column to datetime format
df["Order Date"] = pd.to_datetime(df["Order Date"])
```

```python
df["Year"] = df["Order Date"].dt.year.astype("Int64")
df["Month_name"] = df["Order Date"].dt.month_name()
df["Day"] = df["Order Date"].dt.day.astype("Int64")

df["Delivered"] = df["Shipping Status"] == "Delivered"
```

```python
df.dtypes
```

```python
df["Age"].unique()
```

```python
bins = [0,20,30,40,50,60,100]
labels = ["<20","20-29","30-39","40-49","50-59","60+"]

df["AgeGroup"] = pd.cut(df["Age"], bins=bins, labels=labels)
```

```python
df.to_csv("../data/clean/clean_data.csv", index=False)
```
