```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("../data/clean/clean_data.csv")
```

```python
# Convert "Date" column to datetime format
df["Order Date"] = pd.to_datetime(df["Order Date"])
```

```python
df["Year"] = df["Order Date"].dt.year.astype("Int64")
df["Month_name"] = df["Order Date"].dt.month_name()
df["Day"] = df["Order Date"].dt.day.astype("Int64")

df["Delivered"] = df["Shipping Status"] == "Delivered"
```

- ผมสร้าง Column ใหม่ คือ (ปี, ชื่อเดือน, วันที่) เพื่อที่จะไช้ในการ วิเคราะห์ เกี่ยวกับ สิ้่งที่เกี่ยวข้องกับวันเวลา เช่น ยอดOrder ต่อเดือน, Deliveryต่อเดือนได้ง่ายขึ้น

```python
bins = [0,20,30,40,50,60,100]
labels = ["<20","20-29","30-39","40-49","50-59","60+"]

df["AgeGroup"] = pd.cut(df["Age"], bins=bins, labels=labels)
```

```python
df.columns
```

```python
df["Revenue"] = df["Total Price"] * (df["Delivered"]==True)
```

```python
df["OrderID"] = (
    df["Customer ID"].astype(str)
    + "_"
    + df["Order Date"].astype(str)
)
```
