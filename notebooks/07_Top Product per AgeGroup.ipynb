```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("../data/clean/clean_data.csv")
```

```python
top_age_per_product = (
    df[(df["Year"] == 2023) & (df["Delivered"] == True)]
    .groupby(["Product Name", "AgeGroup"])["Quantity"]
    .sum()
)

top_age_per_product = top_age_per_product.loc[
    top_age_per_product.groupby(level=0).idxmax()
]
```

```python
print(top_age_per_product.sort_values(ascending=False))
```
