### การวิเคราะห์ ยอด Deleivery หรือ Orders ตามช่วงวัย ปี2023

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("../data/clean/clean_data.csv")
```

#### 1 . สินค้าที่ถูก จัดส่ง ที่มากที่สุด แต่ละช่วงวัย ของ Data set

```python
top_age_per_product_delivered = (
    df[(df["Year"] == 2023) & (df["Delivered"] == True)]
    .groupby(["Product Name", "AgeGroup"])["Quantity"]
    .sum()
)

top_age_per_product_delivered = top_age_per_product_delivered.loc[
    top_age_per_product_delivered.groupby(level=0).idxmax()
]
```

```python
print(top_age_per_product_delivered.sort_values(ascending=False))
```

#### 2 . สินค้าที่ถูก สั่ง ที่มากที่สุด แต่ละช่วงวัย ของ Data set

```python
top_age_per_product_orders = (
    df[df["Year"] == 2023]
    .groupby(["Product Name", "AgeGroup"])["Quantity"]
    .sum()
)

top_age_per_product_orders = top_age_per_product_orders.loc[
    top_age_per_product_orders.groupby(level=0).idxmax()
]
```

```python
print(top_age_per_product_orders.sort_values(ascending=False))
```
