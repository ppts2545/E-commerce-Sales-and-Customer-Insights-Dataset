```python
import pandas as pd
import numpy as np

df = pd.read_csv('../data/raw/realistic_e_commerce_sales_data.csv')
```

```python
# Set display options to show all columns and adjust width for better readability
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)

print(df.head())
```

```python
print(df.info())

print("\n")

print(df.describe())
```

```python
#Check for missing values of Each column in percent
print((df.isnull().mean() * 100).sort_values(ascending=False))

print("\n")

#Check for duplicated rows
print("Data duplicate:", df.duplicated().sum())
```

```python

#Convert 'Age' column to numeric, coercing errors to NaN, and then to integer type
df["Age"] = pd.to_numeric(df["Age"], errors="coerce").astype("Int64")
print(df["Age"].unique())
print("\n   ")

#Check unique values in 'Category' column
print(df["Category"].unique())
print("\n   ")

print("All Shipping Status :", df["Shipping Status"].unique())
```

```python
#Check unique values in 'Shipping Fee' column
print(np.sort(df["Shipping Fee"].unique()))
```

```python
print(np.sort(df["Product Name"].unique()))  
```

```python
#Cleaning Data

#Group 'Age' into categories and handle missing values
df["Age_group"] = pd.cut(df["Age"], bins=[18,25,35,50,100])
df["Age_group"] = df["Age_group"].cat.add_categories("Unknown")
df["Age_group"] = df["Age_group"].fillna("Unknown")

#Handle missing values in 'Region' and 'Shipping Status' columns by filling with 'Unknown'
df = df.fillna({"Region": "Unknown"})
df = df.fillna({"Shipping Status": "Unknown"})

```

```python
#Check for missing values of Each column in percent after cleaning
print((df.isnull().mean() * 100).sort_values(ascending=False))
```

```python
df.to_csv("../data/clean/clean_data.csv", index=False)
```
